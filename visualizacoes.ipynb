## Analisando as tendencias de imigração do Brasil em um deterinado período

```python
df.head(5)
```

```python
# Trocando nome da coluna country para pais

df = df.rename(columns={'Country':'País'})
```

```python
# Definindo a coluna país como index do nosso df

df.set_index('País',inplace=True)
```

```python
df.head(5)
```

```python
# Armazenando intervalo de tempo

anos = list(map(str, range(1980,2014)))
```

```python
# Pegando dados apenas do Brasil, relativo ao intervalo de tempo

brasil = df.loc['Brazil', anos]
```

```python
brasil
```

```python
#Criando um dicionario para nomear cada coluna, com o ano e quantidade de imigrantes

brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head(5)
```

## Criando gráfico usando Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

#### Melhorando visualizacao

```python
plt.figure(figsize=(10,5))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(anos[::5])
plt.title('Imigracao do Brasil para o Canada ')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes');
```

## Criando uma figura

```python
#fig cria a figura
#ax é o eixo

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5));
```

## Criando subplots em uma direção

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))

#primeiro gráfico
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980-2013 ')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

#segundo gráfico
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980-2013 ')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid();
```

## Criando subpots em duas direções

#### Analisando os 4 maiores países da América do Sul

```python
# Criando uma figura com 4 graficos (4 subplots)
fig, axs = plt.subplots(2,2, figsize=(10,5))
# Mudando espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.5)

# 1 grafico
axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brasil')

# 2 grafico
axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

# 3 grafico
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

# 4 grafico
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru');

# Ajustando o eixo x dos gráficos
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    
# Adicionando rótulo
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
```

## Deixando todos os gráficos com valores mínimos e máximos

```python
# Criando uma figura com 4 graficos (4 subplots)
fig, axs = plt.subplots(2,2, figsize=(10,5))
# Mudando espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.5)
# Adicionando título principal dos subplots
fig.suptitle('Imigração dos quatro maiore píses da América do Sul para o Canada\n1980-2013')

# 1 grafico
axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brasil')

# 2 grafico
axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

# 3 grafico
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

# 4 grafico
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru');

# Ajustando o eixo x dos gráficos
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    
# Adicionando rótulo
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
    
# Valor míminimo do eixo y
ymin = 0
# Valor máximo do eixo y
ymax = 7000

# Instrução for para ajustar os valores
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
```
