## Analisando as tendencias de imigração do Brasil em um deterinado período

```python
df.head(5)
```

```python
# Trocando nome da coluna country para pais

df = df.rename(columns={'Country':'País'})
```

```python
# Definindo a coluna país como index do nosso df

df.set_index('País',inplace=True)
```

```python
df.head(5)
```

```python
# Armazenando intervalo de tempo

anos = list(map(str, range(1980,2014)))
```

```python
# Pegando dados apenas do Brasil, relativo ao intervalo de tempo

brasil = df.loc['Brazil', anos]
```

```python
brasil
```

```python
#Criando um dicionario para nomear cada coluna, com o ano e quantidade de imigrantes

brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head(5)
```

## Criando gráfico usando Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

#### Melhorando visualizacao

```python
plt.figure(figsize=(10,5))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(anos[::5])
plt.title('Imigracao do Brasil para o Canada ')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes');
```

## Criando uma figura

```python
#fig cria a figura
#ax é o eixo

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5));
```

## Criando subplots em uma direção

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))

#primeiro gráfico
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980-2013 ')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

#segundo gráfico
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980-2013 ')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid();
```

## Criando subpots em duas direções

#### Analisando os 4 maiores países da América do Sul

```python
# Criando uma figura com 4 graficos (4 subplots)
fig, axs = plt.subplots(2,2, figsize=(10,5))
# Mudando espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.5)

# 1 grafico
axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brasil')

# 2 grafico
axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

# 3 grafico
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

# 4 grafico
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru');

# Ajustando o eixo x dos gráficos
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    
# Adicionando rótulo
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
```

## Deixando todos os gráficos com valores mínimos e máximos

```python
# Criando uma figura com 4 graficos (4 subplots)
fig, axs = plt.subplots(2,2, figsize=(10,5))
# Mudando espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.5)
# Adicionando título principal dos subplots
fig.suptitle('Imigração dos quatro maiore píses da América do Sul para o Canada\n1980-2013')

# 1 grafico
axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brasil')

# 2 grafico
axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

# 3 grafico
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

# 4 grafico
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru');

# Ajustando o eixo x dos gráficos
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    
# Adicionando rótulo
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
    
# Valor míminimo do eixo y
ymin = 0
# Valor máximo do eixo y
ymax = 7000

# Instrução for para ajustar os valores
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
```

#### Selecionando paises apenas da america do sul

```python
america_sul = df.query('Region == "South America"')
```

```python
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue',
         'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(8,5))

ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12);
```

## Ordenando o gráfico

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brazil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(8,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canada no período de 1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12);
```

## Adicionando anotações

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brazil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(8,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canada no período de 1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Adicionando numero total de pessoas que imigram dos seus paises para o canada na barra
for i, var in enumerate(america_sul_sorted['Total']):
    ax.text(var+20,i, str(var), color='black', fontsize=10, ha='left', va='center')
    
# Removendo borda da caixa
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
# Removendo valores do eixo X
ax.get_xaxis().set_visible(False)
# Removendo os ticks '-'
ax.tick_params(axis='both', which='both', length=0);
```
