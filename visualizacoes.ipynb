## Analisando as tendencias de imigração do Brasil em um deterinado período

```python
df.head(5)
```

```python
# Trocando nome da coluna country para pais

df = df.rename(columns={'Country':'País'})
```

```python
# Definindo a coluna país como index do nosso df

df.set_index('País',inplace=True)
```

```python
df.head(5)
```

```python
# Armazenando intervalo de tempo

anos = list(map(str, range(1980,2014)))
```

```python
# Pegando dados apenas do Brasil, relativo ao intervalo de tempo

brasil = df.loc['Brazil', anos]
```

```python
brasil
```

```python
#Criando um dicionario para nomear cada coluna, com o ano e quantidade de imigrantes

brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head(5)
```

## Criando gráfico usando Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

#### Melhorando visualizacao

```python
plt.figure(figsize=(10,5))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(anos[::5])
plt.title('Imigracao do Brasil para o Canada ')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes');
```

## Criando uma figura

```python
#fig cria a figura
#ax é o eixo

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5));
```

#### Selecionando paises apenas da america do sul

```python
america_sul = df.query('Region == "South America"')
```

```python
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue',
         'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(8,5))

ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12);
```

## Ordenando o gráfico

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brazil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(8,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canada no período de 1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12);
```

## Adicionando anotações

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brazil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(8,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canada no período de 1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Adicionando numero total de pessoas que imigram dos seus paises para o canada na barra
for i, var in enumerate(america_sul_sorted['Total']):
    ax.text(var+20,i, str(var), color='black', fontsize=10, ha='left', va='center')
    
# Removendo borda da caixa
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
# Removendo valores do eixo X
ax.get_xaxis().set_visible(False)
# Removendo os ticks '-'
ax.tick_params(axis='both', which='both', length=0);
```

## Alterando as cores com paletas

```python
#funcao para escoler paleta de cores
def gerar_grafico_paleta(pallete):
    fig, ax = plt.subplots()
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=pallete)
    
    ax.set_title('Paises com maior imigracao para o canada\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('');
```

#### Paleta de cores sequenciais, bastante usada para respresentar sequencias de valores onde as cores mais claras vao indicar os valores mais baixo e as mais escuras os valores mais alto

```python
gerar_grafico_paleta('Blues')
```

#### Invertendo as cores

```python
gerar_grafico_paleta('Blues_r')
```

#### Paleta divergente

```python
gerar_grafico_paleta('rocket')
```

```python
gerar_grafico_paleta('coolwarm')
```

#### Paleta categorica

```python
gerar_grafico_paleta('tab10')
```
